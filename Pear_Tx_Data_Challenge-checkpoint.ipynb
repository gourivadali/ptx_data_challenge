## Modeling option 1: flavor popularity overall

Given the data that you have, how could you model which flavors are most popular overall? 

The challenge here is that not all of the flavors have been compared to every other flavor at the samelocation
and time. Additionally, you’ll have to come up with a way to quantify how popular a flavor is “locally”
(i.e. at one single vending station), and then extend this across locations to a global approach.

Output a leaderboard of flavors, and come up with a reasonable recommendation for which flavors
should be cut and why.

Can you think of a way to recommend new flavors that haven’t been created yet?

```r
# Data Wrangling to create the Modeling dataset

mat.flavor <- station.flavor.data %>% ungroup() %>%  select(id,station.date,flavor_name, proportion_of_flavors_dispensed_per_day) %>%  mutate(name = "id") %>% select(name, id,station.date, everything()) %>% tidyr::unite(.,"name_id",1:2,sep = "_") %>% mutate(station.id = name_id) %>% tidyr::unite(.,"date",1:2,sep = "_") %>%  distinct(.keep_all = TRUE) %>% tidyr::spread(.,flavor_name,proportion_of_flavors_dispensed_per_day,fill = 0)

rownames(mat.flavor) <- mat.flavor$date

mat.flavor$date <- NULL

mat.flavor$station.id <- as.factor(mat.flavor$station.id)
```

### Generating multiple models

```r
set.seed(187)

seeds <- floor(runif(10, min=0, max=101))

tr_control <-
  trainControl(
    method = "repeatedcv",
    number = 5,
    repeats = 5,
    savePredictions = TRUE,
    classProbs = TRUE,
    verboseIter = TRUE
  )

tuneGrid <- expand.grid(.mtry = 9)

rf_dat <- function(data, seed) {
  set.seed(seed)
  sample = sample(nrow(data), 0.7 * nrow(data), replace = FALSE)
  return(sample)
}

rf_model <- function(data, seed) {
  sample <- rf_dat(data, seed)
  trainset = data[sample, ]
  testset  = data[-sample, ]
  model <- train(
    station.id ~ .,
    trainset,
    method = "rf",
    metric = "Accuracy",
    tuneGrid = tuneGrid,
    trControl = tr_control,
    importance = TRUE,
    nodesize = 7,
    ntree = 501,
    maxnodes = 10
  )
  pred_flavor <- predict(model, testset)
  conf_mat <- confusionMatrix(pred_flavor, testset$station.id)
  varimp <- varImp(model)
  pimp.varimp <-
    PIMP(trainset[, -1],
         trainset$station.id,
         model$finalModel,
         S = 100,
         parallel = FALSE)
  pimp.test <- PimpTest(pimp.varimp)
  result <-
    list(
      model = model,
      pred = pred_flavor,
      confmat = conf_mat,
      varimp = varimp,
      pimpvarimp = pimp.varimp,
      pimptest = pimp.test$pvalue
    )
  return(result)
}

list_of_models <- vector(mode = "list", length = 10)

for (i in 1:length(seeds)) {
  list_of_models[[i]] <-rf_model(mat.flavor, seeds[i])
}
```

### Picking the predictors frequently occuring as imp. variables in all the RF models generated using the PIMP algorithm

```r
mods  = list()

melt_join_dat <- function(data, i) {
  #' Reshapes data to obtain permutational
  #' variable importance and p-value for
  #' the predictors across models.
  #'
  #' Given a model object and number
  #' returns two lists. Uses reshape2 that changes model to a matrix.
  #' @param data list of model objects
  #' @param i an integer indicating model iteration.
  require("dplyr")
  require("reshape2")
  x <- melt(as.matrix(data[[i]]$pimpvarimp$VarImp))
  y <- melt(as.matrix(data[[i]]$pimptest))
  z <- melt(as.matrix(data[[i]]$pimpvarimp$PerVarImp))
  colnames(x)[3] <- "varimp_value"
  colnames(y)[3] <- "pvalue"
  z <-
    z %>% group_by(Var1) %>% mutate(mean_permuted_value = mean(value)) %>% dplyr::select(Var1, mean_permuted_value) %>% distinct(.keep_all = TRUE)
  x$Var1 <- as.character(x$Var1)
  y$Var1 <- as.character(y$Var1)
  z$Var1 <- as.character(z$Var1)
  imp.var <-
    left_join(x, y, by = "Var1") %>% left_join(., z, by = "Var1") %>% dplyr::select(-Var2.x,-Var2.y)
  colnames(imp.var)[1] <- "Flavours"
  imp.var <-
    imp.var %>% mutate(Freq = if_else((as.numeric(pvalue) < 0.05) &
                                       (as.numeric(
                                         mean_permuted_value
                                       ) > 0), 1, 0))
  colnames(imp.var)[2] <- paste0("varimp_value_Model_", i)
  colnames(imp.var)[3] <- paste0("pvalue_Model_", i)
  colnames(imp.var)[4] <- paste0("mean_permuted_value_Model_", i)
  colnames(imp.var)[5] <- paste0("Freq_Model_", i)
  return(imp.var)
}

for (i in 1:length(list_of_models)) {
  mods[[i]] <- melt_join_dat(list_of_models, i)
}

library(purrr)
df_flavors <- mods %>% reduce(full_join, by = "Flavours")

df_imp_flavors <-
  df_flavors %>% dplyr::select(Flavours,Freq_Model_1,Freq_Model_2,Freq_Model_3,Freq_Model_4,Freq_Model_5,Freq_Model_6,Freq_Model_7,Freq_Model_8,Freq_Model_9,Freq_Model_10) %>% mutate(Flavor_Frequency = apply(.[2:11], MARGIN = 1, FUN = sum))
```

### 2.3  Extracting flavors that were predicted to be significantly important across different models.

```r
## Generating a flavour frequency Ratio
df_imp_flavors <-
  df_imp_flavors %>% mutate(Frequency_ratio = Flavor_Frequency / 10) %>% filter(Frequency_ratio > 0) %>%  dplyr::select(Flavours, Flavor_Frequency, Frequency_ratio)

# ggplot for the important variables that have a p-value < 0.05 & mean_pervarimp > 0 from all the RF models
ggplot(df_imp_flavors, aes(x = reorder(Flavours, Flavor_Frequency), y = Flavor_Frequency)) + geom_bar(stat = "identity", fill = "indianred") + coord_flip() + theme_bw() + ggtitle("Important Flavor features from RF Classification Model") + labs(x = "", y = "Overall Flavor Frequency in RF Models")
```
